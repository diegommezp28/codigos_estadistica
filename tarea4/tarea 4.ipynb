## 4

```python
observaciones = np.array([70, 105, 98, 125, 90, 100, 95, 117])
print(observaciones)
```

### 4.c

```python
npi = 800*1/8
np.random.seed(0)
Q7 = np.array([])
for i in range(0, 1000):
    sample = multinomial.rvs(n=1, p=[1/8]*8, size=800).sum(axis=0)
    q7_i = np.sum(((sample-npi)**2)/npi)
    Q7 = np.append(Q7, q7_i)

fig, ax = plt.subplots(figsize=(13,5))
plt.hist(Q7, 30, alpha=0.1, density=True, color='b')
df = 7 #Grados de libertad
x = np.linspace(chi2.ppf(0.01, df), chi2.ppf(0.99, df), 100)
ax.plot(x, chi2.pdf(x, df), 'r-', lw=2, alpha=0.6, label=f'$\chi^2$({df}) pdf')
plt.xlabel("Q7")
plt.ylabel("Densidad")
plt.legend()
plt.show()
```

### 4.d

```python
X = []
Y = []
for i in range (1, 31):
    p = i / 30
    xp = np.quantile(Q7, p)
    yp = chi2.ppf(p, df)
    X.append(xp)
    Y.append(yp)

x = np.linspace(0,15,100)
fig, ax = plt.subplots(figsize=(15,5))
ax.scatter(x=X, y=Y, c='r', label='Muestras')

ax.plot(x,x, 'b', label='$f(x)=x$')
plt.xlabel("p-esimo percentil empírico")
plt.ylabel(f"p-esimo percentil teórico de $\chi^2$({df})")
plt.legend()
plt.show()
```

### 4.e

```python
npi = 100
estadístico = np.sum(((observaciones-npi)**2)/npi)
pvalue = 1-chi2.cdf(estadístico, df)

print(f'Observaciones hechas: {observaciones}')
print(f'Estadístico Q_7 observado: {estadístico}')
print(f'P-value encontrado: {pvalue}')
```
